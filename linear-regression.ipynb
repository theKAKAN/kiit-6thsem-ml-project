```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the CSV data into a DataFrame
df = pd.read_csv("data/clean_data.csv", low_memory=False)
```

```python
df
```

```python
from datetime import datetime
import time
# Convert to unix timestamp
df['ts'] = pd.to_datetime(df['ts'], format='%Y-%m-%d %H:%M:%S')
df['ts'] = df['ts'].apply(lambda x: time.mktime(x.timetuple()))
df
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df["ts"], df["number"], test_size=0.2, random_state=42)
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test.values.reshape(-1, 1))
```

```python

# Print the results
print("Coefficients:", model.coef_[0])
print("Intercept:", model.intercept_)
print("Mean squared error:", model.score(X_test.values.reshape(-1, 1), y_test))
```
